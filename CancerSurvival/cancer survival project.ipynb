Q-    Which year does most of the operations occured ? 

```python
 bcancer_df['Years_of_operation'].plot(kind='hist', figsize=(15,7),color='brown',fontsize=15,yticks=np.arange(0,70,3))
plt.xlabel('Years_of_operation',fontsize=15)
plt.ylabel('Frequency',fontsize=15)
plt.title('Histogram showing the distribution of years of operation')
```

```python
(bcancer_df['Years_of_operation'].value_counts() / len(bcancer_df)).plot(kind='bar', figsize=(12,7),colormap='magma',fontsize=15)
plt.xlabel('Years_of_operation',fontsize=12)
plt.ylabel('Frequency',fontsize=12)
plt.title('Histogram showing the distribution of years of operation')
plt.show()
 
```

1958 is the year when most of the operations occoured.

Q-2 At what age there were chances of survival ?


```python
bcancer_df['Age'].plot(kind='kde', figsize=(15,7), color='blue')

plt.xlabel('Age', fontsize=15)
plt.ylabel('Status', fontsize=15)
plt.title=('KDE plot showing the survival rate in Age')
plt.show()
```

The data in Age column is somewhat following a normal distribution as shape of the histogram is similar to the bell-shaped curve of the normal distribution. 

The range of age survived most is from 40-65 age

Q- In which years of operation higher positive axillary nodes where found out ?

```python
bcancer_df.plot(kind='hexbin',x='Pos_axillary_nodes',y='Years_of_operation',figsize=(15,7),gridsize=25,fontsize=15,colormap='Reds')
plt.show()
```

In 1958,1960,1963,1964,1965,1966 where the higher positive axillary nodes found out. The highest positive axillary nodes was found out in the year 1958.

Q- Is there any  survival chances even if pos_axillary nodes are many ? 

```python
bcancer_df.groupby(['Pos_axillary_nodes'])['Status'].mean().sort_values().plot(kind='line', figsize=(15,7),color='blue',fontsize=13,linestyle='-.')
plt.xlabel('Pos_axillary_nodes',fontsize=12)
plt.ylabel('Status',fontsize=12)
plt.title=('Line Chart showing the mean values of pos_axillary_nodes with status')
plt.show()
```

As the Positive axillary nodes increases there are survival chance as well as much more chances of dead for the patient with 5 years.
Positive axillary nodes does not depend on the survival state.


Q-How is the co-relation between the data?

```python
corr_mat=bcancer_df.corr()
plt.figure(figsize=(10,8))
sns.heatmap(corr_mat,annot=True,cmap='viridis')
plt.title=("The co-relation of data")
```

The data is negatively correlated as the lowest 0.0 is covered all over the chart.Only pos_axillary_nodes and Status are slightly co-related with each other.

Q- Which colums are co-related?

```python
sns.pairplot(data=bcancer_df[['Age','Years_of_operation','Pos_axillary_nodes','Status']],size=2.5,diag_kind='kde')
```

The diagonal elements in the pairplot show the frequency distribution of that column 

There  is a corelation in Years of Operation and Positive axillary nodes are negatively co-related.

Q- Below  what number of Pos_axillary_nodes are women more likely to survive?

```python
bcancer_df.groupby(['Pos_axillary_nodes'])['Status'].mean().sort_values().plot(kind='line',figsize=(15,7), color='green',linestyle='--')
plt.xlabel('Pos_axillary_nodes',fontsize=12)
plt.ylabel('Status',fontsize=12)
plt.title=('Line Chart showing the mean values of pos_axillary_nodes with status')
plt.show()
```

In the range of 16 to 46 are the chances of survival.Below 46 there are chances of survival.
